```python
print(f"../{test['img_path'][5]}")
image_number = 5
cv_image = cv2.imread(f"../{test['img_path'][image_number]}")
for testArray in test['pred_trajs'][image_number]:
    for anotherTest in testArray: 
        # rect = cv2.boundingRect(testArray)
        x, y, w, h = int(anotherTest[0]), int(anotherTest[1]), int(anotherTest[2]), int(anotherTest[3])
        cv2.rectangle(cv_image, (x, y), (x+w, y+h), (0, 255, 0), 2)


for anotherTest in test['gt_trajs'][image_number]:
    # rect = cv2.boundingRect(testArray)
    x, y, w, h = int(anotherTest[0]), int(anotherTest[1]), int(anotherTest[2]), int(anotherTest[3])
    cv2.rectangle(cv_image, (x, y), (x+w, y+h), (0, 0, 255), 2)

RGB_img = cv2.cvtColor(cv_image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
print()
cv_image = cv2.imread(f"../{test['img_path'][200]}")
for testArray in test['pred_trajs'][200]:
    for anotherTest in testArray:
        # rect = cv2.boundingRect(testArray)
        x, y, w, h = int(anotherTest[0]), int(anotherTest[1]), int(anotherTest[2]/8), int(anotherTest[3]/8)
        cv2.rectangle(cv_image, (x, y), (x+w, y+h), (0, 255, 0), 2)

RGB_img = cv2.cvtColor(cv_image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
